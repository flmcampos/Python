```python
import requests
from bs4 import BeautifulSoup
```

```python
#Used headers/agent because the request was timed out and asking for an agent. 
#Using following code we can fake the agent.
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36 OPR/77.0.4054.277'}
response = requests.get("https://www.zomato.com/bangalore/top-restaurantes", headers=headers)
```

```python
content=response.content
soup=BeautifulSoup(content,"html.parser")
```

```python
print(soup.prettify())
```
